```python
# 1
def analyze(likes):
  total=0
  for i in likes:
    total+=i
  if total>=1000:
    status = 'Viral Post'
  else:
    status = 'Normal Engagement'
  print('Total Likes:',total)
  print('Post Status:',status)

likes=[100,200,300,450]
analyze(likes)
```

```python
# 2
def level(stock):
  if stock < 10:
    status = "Low Stock Alert"
  else:
    status = "Stock Sufficient"
  print("Medicine Stock:", stock)
  print("Status:", status)
level(6)
```

```python
# 3
def checker(data, required):
  total = 0
  for i in data:
    total += i
  average = total / len(data)
  if average >= required:
    status = 'Adequate Rainfall'
  else:
    status = 'Inadequate Rainfall'
  print('Average Rainfall:', average)
  print('Rainfall Status:', status)
list = [60, 70, 80, 78]
required = 70
checker(list, required)
```

```python
# 4
def check(users):
  unq = set(users)
  if len(users) != len(unq):
    result = 'Yes'
  else:
    result = 'No'

  print("Duplicate Accounts Found:", result)
users = ['virat','warner','kane','virat']
check(users)
```

```python
# 5
def eligibility(age):
  if age >= 18:
    status = 'Eligible'
  else:
    status = 'Not Eligible'
  print('Patient Age:', age)
  print('Eligibility Status:', status)
eligibility(19)
```

```python
# 6
def premium(prices):
  list = []
  for i in prices:
    if i>2000:
      list.append(i)
  print('Premium Crops:',list)
prices = [1990, 2500, 1800, 3200, 999]
premium(prices)
```

```python
# 7
def system_health(errors):
  if errors == 0:
    status = 'Healthy'
  elif errors <= 5:
    status = 'Minor Issues'
  else:
    status = 'Critical Issues'
  print('Error Count:', errors)
  print('System Status:', status)
system_health(6)
```

```python
# 8
def transaction(amount):
  daily_limit = 50000
  if amount <= daily_limit:
    status = 'Approved'
  else:
    status = 'Rejected'
  print('Transaction Amount:', amount)
  print('Transaction Status:', status)
transaction(50100)
```

```python
# 9
def check_attendance(list):
  total = len(list)
  attended = 0
  for status in list:
    if status == 1:
      attended += 1
  percentage = (attended/total)*100
  if percentage >= 75:
    result = 'Eligible'
  else:
    result = 'Not Eligible'
  print('Attendance Percentage:', percentage)
  print('Exam Eligibility:', result)
list = [1, 1, 1, 0, 1]
check_attendance(list)
```

```python
# 10
def electricity_bill(units):
  bill = 0
  if units <= 100:
    bill = units * 3
  elif units <= 200:
    bill = (100 * 3) + (units - 100) * 5
  else:
    bill = (100 * 3) + (100 * 5) + (units - 200) * 7
  if bill < 500:
    usage = 'Low Usage'
  elif bill <= 1500:
    usage = 'Moderate Usage'
  else:
    usage = 'High Usage'
  print('Units Consumed:', units)
  print('Total Bill:', bill)
  print('Usage Level:', usage)
electricity_bill(250)
```
